```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = application_df["IS_SUCCESSFUL"].values
X = application_df.drop(["IS_SUCCESSFUL"], axis=1).values
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

**TRIAL 0:**

```python
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 10
hidden_nodes_layer2 = 8
hidden_nodes_layer3= 6

nn_model_trial0 = tf.keras.models.Sequential()

# First hidden layer
nn_model_trial0.add(tf.keras.layers.Dense(units=hidden_nodes_layer1,
             input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn_model_trial0.add(tf.keras.layers.Dense(
    units=hidden_nodes_layer2, activation="sigmoid"))

# Third hidden layer
nn_model_trial0.add(tf.keras.layers.Dense(
    units=hidden_nodes_layer3, activation="sigmoid"))

# Output layer
nn_model_trial0.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))


# Check the structure of the model
nn_model_trial0.summary()
```

```python
# Compile the model
nn_model_trial0.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model_trial0 = nn_model_trial0.fit(X_train_scaled, y_train, epochs=30)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model_trial0.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
import keras_tuner as kt
import tensorflow as tf
from tensorflow.keras.layers import Input

# Redefine the model creation function here so it can be used in the run_tuner_search function.
number_input_features = len(X_train_scaled[0])
def create_model(hp):
    nn_model_trial5 = tf.keras.models.Sequential()

    #add input layer
    nn_model_trial5.add(Input(shape=(number_input_features,)))
    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','sigmoid'])

    # Allow kerastuner to decide number of neurons in first layer
    nn_model_trial5.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=10,
        step=2), activation=activation))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 6)):
        nn_model_trial5.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=10,
            step=2),
            activation=activation))

    nn_model_trial5.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model_trial5.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn_model_trial5

# Define the search function
def run_tuner_search(X_train_scaled,y_train, X_test_scaled, y_test):

    # Create the tuner
    tuner = kt.Hyperband(
        create_model,
        objective="val_accuracy",
        max_epochs=20,
        hyperband_iterations=2,
        project_name='AlphabetSoupCharity',
        overwrite=True)

    # Run the kerastuner search for best hyperparameters
    tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))

    # Get best model hyperparameters
    best_hyper = tuner.get_best_hyperparameters(1)[0]
    print(best_hyper.values)

    # Get the best model and return it
    best_model = tuner.get_best_models(1)[0]
    # model_loss, model_accuracy = best_model.evaluate(X_test,y_test,verbose=2)
    # print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
    return best_model, tuner # Return best_model and tuner

# run the search, capture the returned model and tuner
best_model, tuner = run_tuner_search(X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
# Get best model hyperparameters using the tuner variable
best_hyper = tuner.get_best_hyperparameters(1)[0]
print(best_hyper.values)
```

```python
# Evaluate best model against full test data using the best_model variable returned from the function
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
